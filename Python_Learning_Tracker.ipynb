# Python Learning Progress Tracker 📊

Welcome to your interactive Python learning dashboard! This notebook helps you track your progress, set goals, and stay motivated on your Python journey.

Use the interactive widgets below to mark completed topics, take quizzes, and visualize your progress.

## Import Necessary Libraries

Let's start by importing the libraries we'll need for interactivity and data visualization.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import pandas as pd
import matplotlib.pyplot as plt
import random

# Enable matplotlib inline plotting
%matplotlib inline
```

## Define Learning Modules

Here we define the Python learning curriculum with modules and topics.

```python
# Define learning modules and topics
learning_modules = {
    "Basics": ["Variables & Data Types", "Operators", "Input/Output", "Comments"],
    "Control Flow": ["If/Else Statements", "For Loops", "While Loops", "Break & Continue"],
    "Functions": ["Defining Functions", "Parameters & Arguments", "Return Values", "Scope"],
    "Data Structures": ["Lists", "Tuples", "Dictionaries", "Sets"],
    "File Handling": ["Reading Files", "Writing Files", "File Modes", "Error Handling"],
    "OOP": ["Classes & Objects", "Inheritance", "Polymorphism", "Encapsulation"],
    "Modules": ["Importing Modules", "Standard Library", "Creating Modules", "Packages"]
}

# Initialize progress tracking
progress = {module: {topic: False for topic in topics} for module, topics in learning_modules.items()}

# Quiz questions
quiz_questions = {
    "Variables & Data Types": [
        {"question": "What is the data type of: x = 5", "options": ["int", "str", "float", "bool"], "answer": "int"},
        {"question": "Which of these is a string?", "options": ["42", "'42'", "42.0", "True"], "answer": "'42'"}
    ],
    "If/Else Statements": [
        {"question": "What will: if 5 > 3: print('Yes')", "options": ["Yes", "No", "Error", "Nothing"], "answer": "Yes"}
    ]
}
```

## Create Interactive Progress Tracker

Use the checkboxes below to mark topics as completed. Your progress will be saved and visualized.

```python
# Create progress tracker widgets
module_selector = widgets.Dropdown(
    options=list(learning_modules.keys()),
    description='Module:',
    value=list(learning_modules.keys())[0]
)

topic_checkboxes = {}
progress_output = widgets.Output()

def update_checkboxes(change):
    selected_module = change['new']
    progress_output.clear_output()
    
    with progress_output:
        print(f"Topics in {selected_module}:")
        for topic in learning_modules[selected_module]:
            if topic not in topic_checkboxes:
                topic_checkboxes[topic] = widgets.Checkbox(
                    value=progress[selected_module][topic],
                    description=topic
                )
                topic_checkboxes[topic].observe(lambda c, t=topic, m=selected_module: 
                    update_progress(c, t, m), names='value')
            display(topic_checkboxes[topic])

def update_progress(change, topic, module):
    progress[module][topic] = change['new']
    print(f"Updated: {topic} in {module} - {'Completed' if change['new'] else 'Not completed'}")

module_selector.observe(update_checkboxes, names='value')

# Display initial widgets
display(module_selector)
display(progress_output)

# Trigger initial update
update_checkboxes({'new': module_selector.value})
```

## Add Quiz Functionality

Test your knowledge with interactive quizzes for each topic!

```python
# Quiz functionality
quiz_topic_selector = widgets.Dropdown(
    options=list(quiz_questions.keys()),
    description='Quiz Topic:'
)

quiz_button = widgets.Button(description='Start Quiz')
quiz_output = widgets.Output()

def start_quiz(b):
    topic = quiz_topic_selector.value
    questions = quiz_questions.get(topic, [])
    
    if not questions:
        with quiz_output:
            print("No quiz available for this topic yet.")
        return
    
    quiz_output.clear_output()
    with quiz_output:
        print(f"Quiz for {topic}:")
        for i, q in enumerate(questions):
            print(f"\nQuestion {i+1}: {q['question']}")
            print(f"Options: {', '.join(q['options'])}")
            print(f"Answer: {q['answer']}")

quiz_button.on_click(start_quiz)

display(quiz_topic_selector)
display(quiz_button)
display(quiz_output)
```

## Generate Progress Reports

Visualize your learning progress with charts and statistics.

```python
# Progress visualization
def generate_report():
    # Calculate completion percentages
    module_completion = {}
    total_topics = 0
    completed_topics = 0
    
    for module, topics in progress.items():
        module_total = len(topics)
        module_completed = sum(topics.values())
        module_completion[module] = (module_completed / module_total) * 100 if module_total > 0 else 0
        total_topics += module_total
        completed_topics += module_completed
    
    overall_progress = (completed_topics / total_topics) * 100 if total_topics > 0 else 0
    
    # Create visualization
    plt.figure(figsize=(10, 6))
    modules = list(module_completion.keys())
    percentages = list(module_completion.values())
    
    plt.bar(modules, percentages)
    plt.title('Python Learning Progress by Module')
    plt.xlabel('Modules')
    plt.ylabel('Completion (%)')
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, 100)
    
    for i, v in enumerate(percentages):
        plt.text(i, v + 1, f'{v:.1f}%', ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()
    
    print(f"\nOverall Progress: {overall_progress:.1f}% ({completed_topics}/{total_topics} topics completed)")
    
    # Motivational message
    if overall_progress == 100:
        print("🎉 Congratulations! You've completed all topics!")
    elif overall_progress >= 75:
        print("🚀 Great progress! You're almost there!")
    elif overall_progress >= 50:
        print("💪 Halfway there! Keep going!")
    else:
        print("🌱 Just getting started! Every journey begins with a single step!")

# Button to generate report
report_button = widgets.Button(description='Generate Progress Report')
report_output = widgets.Output()

def show_report(b):
    report_output.clear_output()
    with report_output:
        generate_report()

report_button.on_click(show_report)

display(report_button)
display(report_output)
```

## Implement User Interface Elements

Navigate through different sections of your learning tracker using the tabs below.

```python
# Create tabbed interface
tab_contents = ['Progress Tracker', 'Quiz', 'Reports', 'Goals']

tab = widgets.Tab()
tab.children = [
    widgets.VBox([module_selector, progress_output]),
    widgets.VBox([quiz_topic_selector, quiz_button, quiz_output]),
    widgets.VBox([report_button, report_output]),
    widgets.HTML("<h3>Set Your Learning Goals</h3><p>Use this space to write down your learning objectives and track them.</p>")
]

for i, title in enumerate(tab_contents):
    tab.set_title(i, title)

display(tab)

# Motivational quote
print("\n💡 Remember: 'The only way to learn a new programming language is by writing programs in it.' - Dennis Ritchie")
```
